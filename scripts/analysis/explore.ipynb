```python
# Load the playlist_to_genres.json file
playlist_to_genres_path = Path(os.getcwd()) / "../_data" / "playlist_to_genres.json"
with open(playlist_to_genres_path, 'r') as f:
    playlist_to_genres = json.load(f)

# Function to get top 5 genres for a playlist
def get_top_genres(playlist_id):
    if playlist_id in playlist_to_genres:
        genre_counts = playlist_to_genres[playlist_id]['genre_song_counts']
        # Sort genres by count in descending order
        sorted_genres = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)
        # Return top 5 genres
        return sorted_genres[:5]
    return []

# Process each playlist file
for playlist_file in os.listdir(playlists_dir):
    if not playlist_file.endswith('.json'):
        continue
        
    playlist_path = playlists_dir / playlist_file
    with open(playlist_path, 'r') as f:
        playlist_data = json.load(f)
    
    # MODIFIED LINE:
    playlist_id = playlist_data.get('playlist_id') # <-- Was playlist_data.get('id')
    
    # Skip if not in main_playlists
    if playlist_file not in main_playlists:
        continue
    
    print(f"\n{playlist_file}:")
    top_genres = get_top_genres(playlist_id)
    
    if top_genres:
        for genre, count in top_genres:
            print(f"  - {genre}: {count} songs")
    else:
        print("  No genre information available")
```
